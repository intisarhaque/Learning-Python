# Singly Linked Lists

A collection of nodes that form a linear sequence.

Each node has reference to object that is an element of the sequence as well as a reference to the next node of the list.

Going through the node is traversing the list.

Next reference of node known as link/pointer therefor traversing aka link/pointer hopping.

Linked lists have constant time insertions and deletions in any position/comparison. Arrays require O(n) to do same thing.
They can continue to expand without specifying size ahead of time.
But to access element in linked list, takes time O(k) to go from head to the kth element. Arrays have constant time to do so.

## Inserting to Single

Linked list doesn't have a fixed size. Uses space proportional to no. of elements.

To create new element at head of list, make new node, set element as new element, set link to previoius head.

Set new head reference to new head.

To create new element at tail of list, make new node, set element, assign link to None, set next reference of old tail to new node. 

Set new tail reference to new tail.

## Removing from Single

Removing element from head of linked list is reverse opration of inserting a new head element.
To do, point head reference to new head element.

Deleting last node is harder?

```python
class Node(object):
    
    def __init__(self, value):
        
        self.value = value
        self.nextnode = None
    
```

```python
a = Node(1)
b = Node(2)
c = Node(3)
a.nextnode = b
b.nextnode = c
```
